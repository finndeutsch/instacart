```python
# Checking for mixed types data
for col in df_customers.columns.tolist():
  mixed_types = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[mixed_types]) > 0:
    print (col)
```

### 4. Data merge <a name="paragraph3"></a>

```python
# Importing 'orders_products_merged_updated.pkl' and displaying head
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_final.pkl'))
ords_prods_merge.head()
```

```python
# Dropping not needed column 'Unnamed: 0' and checking output
ords_prods_merge = ords_prods_merge.drop('Unnamed: 0', axis=1)
ords_prods_merge.head()
```

```python
# Checking shape of 'ords_prods_merge'
ords_prods_merge.shape
```

```python
# Checking head of 'df_customers'
df_customers.head()
```

```python
# Checking shape of 'ords_prods_merge'
df_customers.shape
```

#### Common column for merging the two data frames is 'user_id'. 

```python
# Checking if data types of columns 'user_id' in data frame 'ords_prods_merge' and data frame 'df_customers' are the same
ords_prods_merge['user_id'].dtype == df_customers['user_id'].dtype
```

#### Data types of the 'user_id' columns in the two data frames are not the same.

```python
# Checking data type of column 'user_id' in data frame 'ords_prods_merge'
ords_prods_merge['user_id'].dtype
```

```python
# Checking data type of column 'user_id' in data frame 'df_customers'
df_customers['user_id'].dtype
```

```python
# Rechanging data type for column 'user_id' (df_customers), since it has to be the same data type as in the other data frame, and integer is most suitable for 'user_id'
df_customers['user_id'] = df_customers['user_id'].astype('int64')
df_customers['user_id'].dtype
```

```python
# Rechecking if data types of columns 'user_id' in data frame 'ords_prods_merge' and data frame 'df_customers' are the same
ords_prods_merge['user_id'].dtype == df_customers['user_id'].dtype
```

#### Data types of the 'user_id' columns in the two data frames are the same now.

```python
# Merging both data frames
ords_prods_customers_merge = ords_prods_merge.merge(df_customers, on = 'user_id')
```
